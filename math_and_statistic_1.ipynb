```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("/content/drive/MyDrive/Major Data File.xlsx")
```

```python
data
```

```python
plt.hist(data['Amount Covered'], bins=8, color='skyblue', edgecolor='black')
plt.title('Histogram of Amount Covered')
plt.xlabel('Amount Covered')
plt.ylabel('Dental Claim Number')
plt.show()
```

```python
sample_mean = np.mean(data['Amount Covered'])
sample_std = np.std(data['Amount Covered'], ddof=1)
```

```python
sample_mean
```

```python
sample_std
```

```python
n=len(data)
```

```python
n
```

```python
degrees_of_freedom = n-1
```

```python
degrees_of_freedom
```

```python
confidence_levels= [0.90,0.95,0.99]
```

```python
for confidence_level in confidence_levels:
    t_score = stats.t.ppf((1 + confidence_level) / 2, df=degrees_of_freedom)
    margin_of_error = t_score * (sample_std / np.sqrt(n))
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
```

```python
print(f"Confidence Interval (for {int(confidence_level*100)}% confidence level): "
          f"({round(confidence_interval[0], 2)}, {round(confidence_interval[1], 2)})")
```
